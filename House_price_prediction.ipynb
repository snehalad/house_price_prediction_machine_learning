### 2. Creating Target variable(y) and Feature variables(X)

```python
# create X, y
X = df_train.drop(['SalePrice'], axis=1)
y = df_train['SalePrice']
```

```python
# list down the categorical col
cat_col = [cname for cname in X.columns if X[cname].dtype=="object" and X[cname].nunique()<10]
print(cat_col)
```

```python
# list down thenumerical col
num_col = [cname for cname in X.columns if X[cname].dtype in ['int64', 'float64']]
print(num_col)
```

**Note:** In our data most columns have null values so we are going to use *SimpleImputer* for that and to handle categorical data we will use *OneHotEncoder*

### 3. Split Data into training & testing

```python
# # split data into train test
# X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

### 4. Pipelines

We're using SimpleImputer for numerical transform And for categorical trasform we're using SimpleImputer and OneHotEncoder

```python
# preprocessing for numeric data
numeric_transformer = SimpleImputer(strategy="constant")

# preprocessing for categorical data
categorical_transform = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])
```

```python
# preprocessing for numeric and categorical data
preprocessor = ColumnTransformer(transformers=[
    ('num', numeric_transformer, num_col),
    ('cat', categorical_transform, cat_col)
])
```

### 5. Create a Model

#### 1 . Linear Regression

```python
X_test = df_test # feature (X)
y_test = df_op['SalePrice'] # label (y)
```

```python
len(X_test), len(y_test)
```

```python
# create model
lr_model = LinearRegression()

# call pipeline
lr_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('model', lr_model)
])

# fit model
lr_pipeline.fit(X, y)

# predict values
lr_preds = lr_pipeline.predict(X_test)

# find score 
lr_score = mean_absolute_error(y_test, lr_preds)
print("Mean Absolute error for Linear Regression:", lr_score)
```

```python
lr_preds
```

#### 2. RandomForestRegressor

```python
# create model
rm_model = RandomForestRegressor(n_estimators=100, random_state=0)

# call pipeline
rm_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('model', rm_model)
])

# fit model
rm_pipeline.fit(X, y)

# predict values
rm_preds = rm_pipeline.predict(X_test)

# find score 
rm_score = mean_absolute_error(y_test, rm_preds)
print("Mean Absolute error for Ramdom Forest Regression:", rm_score)
```
